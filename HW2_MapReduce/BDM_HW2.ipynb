```python
import csv
import sys
```

```python
customers = {} # Product ID -> string
revenues = {} # Product ID -> float
with open('sale.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        customers[row[3]] = customers.get(row[3], set()).union([row[0]])
        revenues[row[3]] = revenues.get(row[3], 0) + float(row[4])
```

```python
print(customers)
print(revenues)
```

```python
with open('HW2_output.csv', 'w') as f:
    writer = csv.writer(f)
    for k,v in customers.items():
        writer.writerow([k, len(v), revenues[k]])
```
